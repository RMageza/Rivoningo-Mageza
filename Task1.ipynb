```python
try:
    import pandas as pd
except ImportError:
    raise ImportError("This script requires pandas. Please install it using 'pip install pandas'.")

try:
    import matplotlib.pyplot as plt
except ImportError:
    raise ImportError("This script requires matplotlib. Please install it using 'pip install matplotlib'.")
```

```python
wines = pd.read_excel("wine.xlsx")
wines
```

```python
def find_intervals(x):
  if 0 <= x and x < 1:
    return '(0,1]'
  elif 1 <= x and x < 2:
    return '(1,2]'
  elif 2 <= x and x < 3:
    return '(2,3]'
  elif 3 <= x and x < 4:
    return '(3,4]'

wines['intervals'] = wines['Flavanoids'].apply(find_intervals)
#int_counts = wines.pivot_table(index='Interval', columns='Class', aggfunc ='size', fill_value=0)
#wines.groupby('intervals', as_index=False).count().plot(kind='bar', x='intervals', y='Class', rot=0, width=0.8)
#print(wines[['intervals', 'Flavanoids','Class']])
```

```python
int_counts = wines.pivot_table(index='intervals', columns='Class', aggfunc ='size', fill_value=0)
print(int_counts)
```

```python
int_counts.plot(kind='bar')
plt.title("Histogram of Wine Data")
plt.xlabel("Intervals")
plt.ylabel("Number of Samples")
plt.show()
```
